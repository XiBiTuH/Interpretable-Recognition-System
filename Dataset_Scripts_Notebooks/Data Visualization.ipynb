```python
# Find the image a given directory
def find(name, path):
    for root, dirs, files in os.walk(path):
        if name in files:
            return os.path.join(root, name)

path_to_watch = os.getcwd() + "\\annotations\\"
```

```python
#Conjuntos de Treino
aux  = pickle.load(open("needed_correct_images.p","rb"))
ks = list(aux.keys())
test_scores_1290 = []

for i in tqdm(range(8000,10251)):
    
    size = (192,256)
    
    im = cv.imread(find(ks[i], path_to_watch))
    im = cv.cvtColor(im, cv.COLOR_BGR2RGB)


    normalized_image = np.zeros(size)
    normalized_image = cv.normalize(im, normalized_image, 0, 255, cv.NORM_MINMAX)
    resize = cv.resize(normalized_image, size, interpolation=cv.INTER_AREA)
    
    X = (np.asarray(resize) - 127.5) / 127.5
    
    #X = ((X * 127.5) + 127.5).astype(int)
    
    X,chars = [np.asarray([X]),np.asarray([aux[ks[i]]])]
    
    k = discriminator.predict([X,chars])
    
    
    test_scores_1290.append(k[0][0])
    

    
    
    
```
